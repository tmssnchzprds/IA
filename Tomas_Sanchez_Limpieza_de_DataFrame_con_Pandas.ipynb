```python
import pandas as pd

#desde csv
ataques = pd.read_excel("shark attack.xlsx")
ataques.head()
```

```python
ataques = ataques.drop(["Unnamed: 21", "Unnamed: 22","Case Number.1","Case Number.2",
                  "href formula"], axis=1)
```

```python
ataques['Fatal (Y/N)'] = ataques['Fatal (Y/N)'].str.strip() #quitamos espacios vacios
ataques['Fatal (Y/N)'] = ataques['Fatal (Y/N)'].fillna('U') #reemplazamos los nulos por U
ataques['Fatal (Y/N)'] = ataques['Fatal (Y/N)'].str.replace('n', 'N', regex=True) #reemplazamos n por N
ataques['Fatal (Y/N)'] = ataques['Fatal (Y/N)'].str.replace('#VALUE!', 'U', regex=True) #reemplazamos #VALUE! por U
ataques['Fatal (Y/N)'] = ataques['Fatal (Y/N)'].str.replace('F', 'N', regex=True) #reemplazamos F por N
ataques['Fatal (Y/N)'] = ataques['Fatal (Y/N)'].str.replace('UNKNOWN', 'U', regex=True) #reemplazamos UNKNOWN por U
ataques.rename(columns={'Fatal (Y/N)': 'Fatal (Y/N/U)'}, inplace=True) # renombramos la columna
```

```python
ataques.columns = ataques.columns.str.strip()
```

# Limpiamos columna Sex

```python
set(ataques['Sex'])
```

```python
ataques['Sex'] = ataques['Sex'].str.strip()
ataques['Sex'] = ataques['Sex'].fillna('U')
ataques['Sex'] = ataques['Sex'].str.replace('.', 'U', regex=True)
ataques['Sex'] = ataques['Sex'].str.replace('1931.03.15', 'U', regex=True)
ataques['Sex'] = ataques['Sex'].str.replace('2014.10.11', 'U', regex=True)
ataques['Sex'] = ataques['Sex'].str.replace('N', 'U', regex=True)
ataques['Sex'] = ataques['Sex'].str.replace('lli', 'U', regex=True)
```

```python
set(ataques['Sex'])
```

# Columna Country

```python
ataques['Country'] = ataques['Country'].str.strip()
ataques['Country'] = ataques['Country'].fillna('UNKNOWN')
ataques['Country'] = ataques['Country'].str.replace('http://sharkattackfile.net/spreadsheets/pdf_directory/2016.05.29.b-Corona.pdf', 'UNKNOWN', regex=True)
ataques['Country'] = ataques['Country'].str.upper()
ataques['Country'] = ataques['Country'].str.replace('?','', regex=True)
```

```python
set(ataques['Country'])
```

# Reemplazar los valores nulos de la columna 'Species' con 'Unknown'

```python
ataques['Species'] = ataques['Species'].fillna('Unknown')
```

# Guarda el resultado en un csv nuevo

```python
ataques.to_csv('ataques-limpio.csv')
```
